## Method 1

### The following code returns two lists: one that indicates the frequency of A and T at each nucleotide postion, and another one to indicate the frequency of G and C at each nucleotide position.

```python
from Bio import AlignIO
align = AlignIO.read('seqs.aligned.fa','fasta')
#print(align)
#print(align[:,3])

cluster1 = align[:3]
cluster1_at = []
cluster1_cg = []
#percent at in cluster 1 at each position

for i in range(2237):
    at = []
    cg = []
    for j in range(len(cluster1[:,i])):
        if cluster1[:,i][j] == 'G' or cluster1[:,i][j] == 'C':
            cg += [1]
        else:
            at += [1]
    cluster1_at += [sum(at)/(sum(at)+sum(cg))]
    cluster1_cg += [sum(cg)/(sum(at)+sum(cg))]

#cluster1_at
#cluster1_cg
#cluster 1 is made of the following sequences: hu.31, hu.14 and hu.32

```

```python
from Bio import SeqIO
from Bio.Seq import Seq
import numpy as np
import matplotlib.pyplot as plt
```

### Using the frequencies lists generated previously for cluster1, 2 and 3 bar graphs were created

```python

fig = plt.figure()
ax = fig.add_subplot(111)


N = 2237
at1 = cluster1_at
cg1 = cluster1_cg

ind = np.arange(N)               
width = 0.35                      

rects1 = ax.bar(ind, at1, width, color= 'blue')

rects2 = ax.bar(ind, cg1, width, color= 'red', bottom=at1)

ax.set_ylim(0,1)
ax.set_ylabel('% of AT or % of CG')
ax.set_title('Position in sequence - Cluster 1')

plt.legend((rects1[0],rects2[0]),('AT','GC'))

plt.show()
```

```python
filename = "seqs.fa"
records = list(SeqIO.parse(filename, "fasta"))

#for cluster one, we have hu.31, hu.14 and hu.32

print(records[7].id)
record_hu14 = records[7].seq
record_hu14 = list(record_hu14)
print(len(record_hu14))

print(records[9].id)
record_hu31 = records[9].seq
record_hu31 = list(record_hu31)
print(len(record_hu31))

print(records[10].id)
record_hu32 = records[10].seq
record_hu32 = list(record_hu32)
print(len(record_hu32))

# Printing the id and length of each sequences

percentage = []
percentage1 = []

for n in range(0,2211):
    
    at_n=0
    tl=3
    
    if record_hu14[n] == 'a':
        at_n =at_n +1
    
    if record_hu31[n] == 'a':
        at_n =at_n +1
    
    if record_hu32[n] == 'a':
        at_n =at_n +1
    
    if record_hu14[n] == 't':
        at_n =at_n +1
   
    if record_hu31[n] == 't':
        at_n =at_n +1
    
    if record_hu32[n] == 't':
        at_n =at_n +1
    
    pt_n=at_n/tl
    pt1_n=1-pt_n
    
    percentage.append(pt_n)
    percentage1.append(pt1_n)
# Calculating the percentage of at/gc

means_at = tuple(percentage)
means_gc = tuple(percentage1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 2211
means_at = tuple(percentage)
means_gc = tuple(percentage1)

fig, ax = plt.subplots()

# create plot
index = np.arange(1,2212,1)
bar_width = 0.3
opacity = 0.8

rects1 = plt.bar(index, means_at, bar_width,
                 alpha=opacity,
                 color='Y', label="AT")

rects1 = plt.bar(index, means_gc, bar_width,
                 alpha=opacity,
                 color='B', bottom=means_at, label="GC")
 
plt.xlabel('Position in Sequence')
plt.ylabel('%AT or %GC')
plt.title('%AT or %GC in different positions')
plt.xticks(index)
plt.axis([0,2500,0,1.0])
ax.xaxis.set_ticks(np.arange(0, 2500, 500))
plt.legend()

 
plt.tight_layout()
plt.show()
```
