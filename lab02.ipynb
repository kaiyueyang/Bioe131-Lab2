```python
from Bio import AlignIO
align = AlignIO.read('seqs1.aligned.fa','fasta')
#print(align)
#print(align[:,3])

cluster1 = align[:3]
cluster1_at = []
cluster1_cg = []
#percent at in cluster 1 at each position

for i in range(2237):
    at = []
    cg = []
    for j in range(len(cluster1[:,i])):
        if cluster1[:,i][j] == 'G' or cluster1[:,i][j] == 'C':
            cg += [1]
        else:
            at += [1]
    cluster1_at += [sum(at)/(sum(at)+sum(cg))]
    cluster1_cg += [sum(cg)/(sum(at)+sum(cg))]

import numpy as np
import matplotlib.pyplot as plt


import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

## the data
N = 2237
at1 = cluster1_at
cg1 = cluster1_cg


## necessary variables
ind = np.arange(N)                # the x locations for the groups
width = 0.35                      # the width of the bars

## the bars
rects1 = ax.bar(ind, at1, width, color= 'blue')

rects2 = ax.bar(ind, cg1, width, color= 'red', bottom=at1)




# axes and labels

ax.set_ylim(0,1)
ax.set_ylabel('% of AT or % of CG')
ax.set_title('Position in sequence - Cluster 1')

plt.legend((rects1[0],rects2[0]),('AT','GC'))

plt.show()
```

There are only three sequences in this cluster, so makes sense that there are not a lot of variations in the percentages
